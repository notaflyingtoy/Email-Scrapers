```python
url ='https://workingforathinktank.co.uk/uk-think-tank-list/'
agent = {"User-Agent":'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'}

page = requests.get(url, headers=agent).content
soup = BeautifulSoup(page, 'html.parser')  
print(soup)
```

```python
thinktank_links = []
thinktanks = []
thinktank_names = soup.find_all("td", {"class":"col-1 odd"})
thinktank_cols = soup.find_all("td", {"class":"col-3 odd"})
for ttank in thinktank_names:
    thinktanks.append(ttank.string)
for tt in thinktank_cols:
    thinktank_links.append(tt.find("a")["href"])
```
